```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
import plotly.graph_objects as go
import plotly.io as pio
pio.renderers.default = "notebook"
import seaborn as sns
plt.style.use('ggplot')
```

```python
df=pd.read_csv('data_used/covid_19_india.csv')
df['Date']=pd.to_datetime(df['Date'],format="%d/%m/%y")
df=df[df['Date']>='2020-04-01']
df_age=pd.read_csv('data_used/AgeGroupDetails.csv')
st_x=[]
conf_y=[]
st_d=[]
for state in df['State/UnionTerritory'].unique():
    st_x.append(df[df['State/UnionTerritory']==state]['Date'])
    conf_y.append(df[df['State/UnionTerritory']==state]['Confirmed'])
    st_d.append(df[df['State/UnionTerritory']==state]['Deaths']) 
def conf():
    plt.figure(figsize=(18,10))
    for i in range(34):
        sns.lineplot(x=st_x[i],y=conf_y[i],legend=False)
    plt.legend(title='States',bbox_to_anchor=(1.01, 1), loc=2, borderaxespad=0., labels=df['State/UnionTerritory'].unique())
    plt.show()
def deaths(state='all'):
    plt.figure(figsize=(18,10))
    for i in range(34):
        sns.lineplot(x=st_x[i],y=st_d[i],legend=False)
    plt.legend(title='States',bbox_to_anchor=(1.01, 1), loc=2, borderaxespad=0., labels=df['State/UnionTerritory'].unique())
    plt.show()
```

## Statewise Confirmed Cases of Covid-19 since April 1st 2020

```python
conf()
```

# Deaths in states since April 1st 2020

```python
deaths()
```
